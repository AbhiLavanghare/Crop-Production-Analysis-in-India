```python
state_area = df.groupby('State_Name')['Area'].sum().sort_values(ascending=False)
state_area
```

```python
# Plotting the area values by state
plt.figure(figsize=(12, 8))
state_area.plot(kind='bar')
plt.title('Total Crop Area by State')
plt.xlabel('State')
plt.ylabel('Total Area')
plt.xticks(rotation=90)
plt.show()
```

From above we can conclude that :

1)Uttar Pradesh is the state with highest agricultural land.

2)Chandigarh is the state with lowest agricultural land.

```python
##Statewise Crop Production
state_production = df.groupby('State_Name')['Production'].sum().sort_values(ascending=False)
state_production
```

```python
plt.figure(figsize=(12, 8))
state_production.plot(kind='bar')
plt.title('Total Crop Production by State')
plt.xlabel('State')
plt.ylabel('Total Production')
plt.xticks(rotation=90)
plt.show()
```

From above we can conclude that :

1) Kerala has highest production although not in top among agricultural area.

2) Chandigarh has lowest production but in sink with its rank in agricultural area.

```python
X = df.drop(columns=['Production'])
y = df['Production']
```

```python
X.head()
```

```python
y.head()
```

```python
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train_num = X_train.select_dtypes(np.number)
X_train_cat = X_train.select_dtypes(object)

X_test_num = X_test.select_dtypes(np.number)
X_test_cat = X_test.select_dtypes(object)
```

```python
X_train_cat.shape, X_test_cat.shape
```

```python
X_train_num.shape, X_test_num.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_num_scaled = scaler.fit_transform(X_train_num)
X_test_num_scaled = scaler.transform(X_test_num)
```

```python
X_train_num_scaled = pd.DataFrame(X_train_num_scaled, columns=X_train_num.columns)
X_test_num_scaled = pd.DataFrame(X_test_num_scaled, columns=X_test_num.columns)
```

```python
X_train_num_scaled.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse=False,handle_unknown='ignore')
X_train_cat_encoded = ohe.fit_transform(X_train_cat)
X_test_cat_encoded = ohe.transform(X_test_cat)
```

```python
X_train_cat_encoded
```

```python
column_names = ohe.get_feature_names_out(X_train_cat.columns)
X_train_cat_encoded = pd.DataFrame(X_train_cat_encoded, columns=column_names)
X_test_cat_encoded = pd.DataFrame(X_test_cat_encoded, columns=column_names)
```

```python
X_test_cat_encoded.head()
```

```python
X_train_processed = pd.concat([X_train_cat_encoded, X_train_num_scaled], axis = 1)
X_test_processed = pd.concat([X_test_cat_encoded, X_test_num_scaled], axis = 1)
```

```python
X_train_processed.shape, X_test_processed.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
```

```python
lr = LinearRegression()
lr.fit(X_train_processed, y_train)
```

```python
y_pred = lr.predict(X_test_processed)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
```

```python
mse, r2, mae
```

```python
knn = KNeighborsRegressor()
knn.fit(X_train_processed, y_train)
```

```python
y_pred = knn.predict(X_test_processed)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
```

```python
mse, r2, mae
```

```python
dt = DecisionTreeRegressor()
dt.fit(X_train_processed, y_train)
```

```python
y_pred = dt.predict(X_test_processed)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
```

```python
mse, r2, mae
```

```python
forest = RandomForestRegressor()
forest.fit(X_train_processed, y_train)
```

```python
y_pred = forest.predict(X_test_processed)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
```
